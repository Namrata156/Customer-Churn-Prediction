#### Create Plots

```python
sns.set_theme(style='whitegrid')
```

```python
# Segment the columns of the dataset into two lists based on the number of unique values each column contains

count_col = []
hist_col = []

for column in df.columns:
    unique_value = df[column].nunique()
    if unique_value <= 20:
        count_col.append(column)
    else:
        hist_col.append(column)
```

##### Count plots for categorical columns

```python
plt.figure(figsize=(12,36))
plt_num=1
for col in count_col:
    plt.subplot(10,2,plt_num)
    sns.countplot(data=df,x=col)
    plt_num+=1
    plt.tight_layout()
```

##### Histograms for numerical columns

```python
plt.figure(figsize=(12,36))
plt_num = 1
for col in hist_col:
    plt.subplot(10,2,plt_num)
    sns.histplot(data=df,x=col,bins=25)
    plt_num+=1
    plt.tight_layout()
```

##### Plot the churn distribution for each categorical variable

```python
plt.figure(figsize=(12,36))
plt_num=1

for col in count_col:
    if df[col].nunique()<=8 and col != 'churn':
        plt.subplot(10,2,plt_num)
        sns.countplot(data=df,x=col,hue='churn')
        plt_num+=1
        plt.tight_layout()
```
